```python
def find_image_corners(image):
    

    objpts = np.zeros((6*9,3), np.float32)
    objpts[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    ret, corners = cv2.findChessboardCorners(gray, (9,6), None)
    

    return ret, corners, objpts
    
    
```

```python
def perspective_transform(image, src, dst):
#     src = np.array([[220,719],[1105, 719],[720, 470],[565, 470]], np.float32)
#     dst = np.array([[50,720],[1230,720],[1230,0],[50,0]],np.float32)
    
    M = cv2.getPerspectiveTransform(src, dst)
    img_size = (image.shape[1],image.shape[0])
    warped = cv2.warpPerspective(image, M, img_size, flags = cv2.INTER_LINEAR)
    return warped
```

```python
def finding_lane_line(img):
    histogram = np.sum(img[img.shape[0]//2:,:], axis=0)
#     plt.plot(histogram)
#     plt.show()
    y_s = 30/720
    x_s = 3.7/700
    out_img = np.dstack((img, img, img))*255
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    
    nwindows = 9
    margin = 100
    minpix = 50
    
    window_height = np.int(img.shape[0]//nwindows)
    nonzero = img.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    
    leftx_current = leftx_base
    rightx_current = rightx_base
    
    left_lane_inds = []
    right_lane_inds = []

    for i in range(1,nwindows+1):
        
        leftx_min = leftx_current-margin
        rightx_min = rightx_current-margin
        leftx_max = leftx_current+margin
        rightx_max = rightx_current+margin
        
        y_min = img.shape[0]-i*window_height
        y_max = img.shape[0]-(i-1)*window_height
        
        rect_img = cv2.rectangle(out_img, (leftx_min, y_min),(leftx_max, y_max), (255, 0, 0), 3)
        rect_img = cv2.rectangle(out_img, (rightx_min, y_min),(rightx_max, y_max), (255, 0, 0), 3)
        
        nonzero_left = ((nonzerox >= leftx_min) & (nonzerox <= leftx_max) & (nonzeroy >= y_min) & (nonzeroy <= y_max)).nonzero()[0]
        nonzero_right = ((nonzerox >= rightx_min) & (nonzerox <= rightx_max) & (nonzeroy >= y_min) & (nonzeroy <= y_max)).nonzero()[0]
        
        left_lane_inds.append(nonzero_left)
        right_lane_inds.append(nonzero_right)
        
        if(len(nonzero_left) > minpix):
            leftx_current = int(np.mean(nonzerox[nonzero_left]))
            
        if(len(nonzero_right) > minpix):
            rightx_current = int(np.mean(nonzerox[nonzero_right]))
            

    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
        
    except:
        pass
    
    
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds]
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    try:
        curvecoeffl = np.polyfit(lefty, leftx, 2)
        curvecoeffr = np.polyfit(righty, rightx, 2)
        
    except:
        curvecoeffl = [0,0,320]
        curvecoeffr = [0,0,920]
    
    ys = np.linspace(0,719,720)
    lx = curvecoeffl[0]*ys**2 + curvecoeffl[1]*ys + curvecoeffl[2]
    rx = curvecoeffr[0]*ys**2 + curvecoeffr[1]*ys + curvecoeffr[2]
    mid = ((lx[-1]+rx[-1])//2 - 1280/2)*3.7/700
    
#     curvecoeffl = np.polyfit(lefty*y_s, leftx*x_s, 2)
#     curvecoeffr = np.polyfit(righty*y_s, rightx*x_s, 2)
    
#     left_curverad = (1+ (2*curvecoeffl[0]*ys[-1] + curvecoeffl[1])**2)**(3/2)/np.absolute(curvecoeffl[0])  ## Implement the calculation of the left line here
#     right_curverad = (1+ (2*curvecoeffr[0]*ys[-1] + curvecoeffr[1])**2)**(3/2)/np.absolute(curvecoeffr[0])
#     print((left_curverad, right_curverad))
    
#     plt.imshow(rect_img)
#     plt.plot(lx, ys, color='yellow')
#     plt.plot(rx, ys, color='yellow')
#     plt.show()
    
    mask_image = np.zeros_like(out_img)
    pts = np.zeros((2*len(ys),2))
    
    for i in range(len(ys)):
        pts[2*i,:] = [lx[i],ys[i]]
        pts[2*i+1,:] = [rx[i],ys[i]]
        
    cv2.fillPoly(mask_image, np.int32([pts]), [0,255,0])
#     plt.imshow(mask_image)
#     plt.show()
    return lx, rx, mask_image, mid
```

```python
def measure_curvature_pixels(leftx, rightx):
    '''
    Calculates the curvature of polynomial functions in pixels.
    '''
    y_s = 30/720
    x_s = 3.7/700
    try:
        ys = np.linspace(0,719,720)

        left_fit = np.polyfit(ys*y_s, leftx*x_s, 2)
        right_fit = np.polyfit(ys*y_s, rightx*x_s, 2)

        left_curverad = ((1 + (2*left_fit[0]*ys*y_s + left_fit[1])**2)**1.5) / np.absolute(2*left_fit[0])  ## Implement the calculation of the left line here
        right_curverad = ((1 + (2*right_fit[0]*ys*y_s + right_fit[1])**2)**1.5) / np.absolute(2*right_fit[0])  ## Implement the calculation of the right line here

    #     x_left = left_fit[0]*(ys[-1]*y_s)**2 + left_fit[1]*ys[-1]*y_s + left_fit[2]
    #     x_right = right_fit[0]*(ys[-1]*y_s)**2 + right_fit[1]*ys[-1]*y_s + right_fit[2]
        return left_curverad[-1], right_curverad[-1]
    
    except:
        return 0,0
```
